```python
from numpy import linspace, sin, cos
x=linspace(0, 10, 1000)
y=sin(x)
plot(x,y)
show()
from pylab import savefig
plot(x,y)
savefig('figura2:Y=sin(x)', format='eps', dpi=1000)
show()
```

```python
from numpy import linspace, sin, cos, tan
from pylab import ylim, xlim, grid, xlabel, ylabel
x=linspace(0, 10, 100)
y=tan(x)
#plot(x,y)
#plot(x,y,'g--')
plot(x,y,'bo')
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
savefig('figura3:Y=tan(x)', format='eps', dpi=1000)
xlabel("eje X")
ylabel("eje Y")
show()
```

```python
from numpy import linspace, sin, cos, tan
from pylab import ylim, xlim, grid, xlabel, ylabel
x=linspace(0, 10, 1000)
y=sin(x)
#plot(x,y)
#plot(x,y,'g--')
plot(x,y,'k-')
y=cos(x)
plot(x,y,'k--')
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
savefig('figura3:Y=tan(x)', format='eps', dpi=1000)
xlabel("eje X")
ylabel("eje Y")
show()
```
